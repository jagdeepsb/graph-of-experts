```python
# visualize some of the dataset
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    img, attr, target = d[sum(d.dataset_lengths[:i])]
    ax[i].imshow(img.permute(1, 2, 0).cpu().numpy())
    ax[i].set_title(f"Target: {target.item()}")
    ax[i].set_xlabel(d.attr_dict[attr.item()])
```
